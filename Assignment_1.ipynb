```python
job_data = { "Job_ID": [1, 2, 3, 4, 5],
 "Role": ["Data Analyst", "Data Scientist", "Business Analyst", "Data Engineer", "ML Engineer"],
 "Company": ["ABC Corp", "XyZ Ltd", "DataWorks", "InfraTech", "AI Labs"],
 "Skills": [
 "Excel SQL Python",
 "Python ML Statistics",
 "Excel SQL PowerBI",
 "Python SQL Spark",
 "Python ML DeepLearning"
 ]}
print(job_data)
```

```python
#1) Using a for loop, print each job posting in a readable format.
n=len(job_data["Job_ID"])
for i in range(n):
    print("ID:", job_data["Job_ID"][i],"| Role:", 
          job_data["Role"][i],"| Company:", 
          job_data["Company"][i],"| Skills:", 
          job_data["Skills"][i])
```

```python
#2Add New Job Postings (interactive)
# Ask the user how many job postings they want to add. For each posting, collect the role,
# company, and skills (skills provided as a single string, space-separated or comma-separated).
n = int(input("How many jobs postings do you want to add: "))

for i in range(n):
    new_id = max(job_data["Job_ID"]) + 1
    role = input("Role: ")
    company = input("Company: ")
    skills = input("Skills: ")

    # maximum no of job_id present in dictionary
    job_data["Job_ID"].append(new_id)
    job_data["Role"].append(role)
    job_data["Company"].append(company)
    job_data["Skills"].append(skills)

print(job_data)
```

```python
# 3)  Clean & Normalize Skills (text processing)
# Apply clean_skills() to every entry in job_data['Skills'] and overwrite the list with cleaned
# versions.
def clean_skills(skillstr):
    skillstr = skillstr.lower()
    skillstr = skillstr.replace(",", " ")
    skillstr = " ".join(skillstr.split())
    return skillstr

job_data['Skills'] = [clean_skills(i) for i in job_data['Skills']]

print(job_data)
```

```python
# 4)Skill Extraction & Frequency
# ● From the cleaned skills list, split each skill string into individual skill tokens.
# ● Build a consolidated list of all skills that contains every skill token across all postings.
# ● Use a dictionary or other structure to compute skill_count = {skill: frequency}.
# ● Print the top 5 most frequent skills (or all if < 5).
skill_count = {}

for i in job_data["Skills"]:
    for j in i.split():
        j = j.lower()

        if j in skill_count:
            skill_count[j] = skill_count[j] + 1
        else:
            skill_count[j] = 1

print("Top skills by frequency:")

sorted_skills = sorted(skill_count.items(), key=lambda x: x[1], reverse=True)

for k in range(min(5, len(sorted_skills))):
    print(sorted_skills[k][0], ":", sorted_skills[k][1])
```

```python
#5)Basic Statistics & Insights
#Calculate and print:
# Total number of job postings
#Number of unique skills
#Jobs that mention "python" (case-insensitive)

print("Total jobs:", len(job_data["Job_ID"]))

a = job_data["Skills"][0].split()

for i in range(1, len(job_data["Skills"])):
    a += job_data["Skills"][i].split()

print("Unique skills:", len(set(a)))

skill = "python"
print("Jobs mentioning Python:")
for i in range(len(job_data["Job_ID"])):
    if skill in job_data["Skills"][i]:
      print(job_data["Role"][i])

```

```python
#6Skill Search Function (functions & conditional logic)
# Create a function, find_jobs_by_skill(skill_query), that returns a list of (Job_ID, Role,
# Company) for postings containing the skill_query token.
# ● The function must normalize the skill_query (lowercase, strip spaces)
# ● Use if/elif/else where appropriate

def find_jobs_by_skill(skill_query):
  skill_query = skill_query.lower().strip ()
  for i in range(len(job_data["Job_ID"])):
    skills = job_data["Skills"][i]
    if skill_query in skills:
      print(job_data["Role"][i])

find_jobs_by_skill("Sql")


```

```python
# 7)  Lambda Filtering & List Comprehension
# Use a lambda function and filter() (or a list comprehension with a lambda-like expression) to
# extract job roles where the skill count (from skill_count) for any skill >= 2 — that is, roles that
# # include at least one relatively common skill.
skill_count = {}
for i in job_data["Skills"]:
    for j in i.split():
        skill_count[j.lower()] = skill_count.get(j.lower(), 0) + 1

result = [
    job_data["Role"][k]
    for k in range(len(job_data["Role"]))
    if any(skill_count[j.lower()] >= 2 for j in job_data["Skills"][k].split())
]

print("Roles with at least one common skill:", result)
```

```python

# 8) Unique Skill Set
# Convert the consolidated skill list into a set and print the unique skills in alphabetical order.

skills = []

for i in job_data["Skills"]:
    skills += i.split()

unique_skills = list(set(skills))
unique_skills.sort()

print("Unique skills:")
print(unique_skills)
```
